```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_scaled = StandardScaler().fit_transform(df_market_data[["price_change_percentage_24h", "price_change_percentage_7d", "price_change_percentage_14d", "price_change_percentage_30d", "price_change_percentage_60d", "price_change_percentage_200d", "price_change_percentage_1y"]])
```

```python
# Create a DataFrame with the scaled data
df_market_scaled = pd.DataFrame(market_scaled, columns=["price_change_percentage_24h", "price_change_percentage_7d", "price_change_percentage_14d", "price_change_percentage_30d", "price_change_percentage_60d", "price_change_percentage_200d", "price_change_percentage_1y"])

# Copy the crypto names from the original data
df_market_scaled["coin_id"] = df_market_data.index

# Set the coinid column as index
df_market_scaled = df_market_scaled.set_index("coin_id")

# Display sample data
df_market_scaled.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(df_market_scaled)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
og_elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow_og = pd.DataFrame(og_elbow_data)

df_elbow_og.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
k_og = df_elbow_og.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k,
    label="Original data"
)

k_og
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** It appears 4 looks like the best value for 'k', as this is where the curve starts to bend.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialise the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the scaled data
model.fit(df_market_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
market_clusters = model.predict(df_market_scaled)

# Print the resulting array of cluster values.
print(market_clusters)
```

```python
# Create a copy of the DataFrame
df_market_predictions = df_market_scaled.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_market_predictions["market_clusters"] = market_clusters

# Display sample data
df_market_predictions.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Colour the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
predict_og = df_market_predictions.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="market_clusters",
    hover_cols = ["coin_id"],
    label = "Original data"
)

predict_og
```

---

### Optimise Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
market_pca = pca.fit_transform(df_market_scaled)

# View the first five rows of the DataFrame. 
market_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 89.5%

```python
# Create a new DataFrame with the PCA data.
pca_market_df = pd.DataFrame(market_pca, columns = ["PCA1", "PCA2", "PCA3"])

# Copy the crypto names from the original data
pca_market_df["coin_id"] = df_market_data.index

# Set the coinid column as index
pca_market_df = pca_market_df.set_index("coin_id")

# Display sample data
pca_market_df.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia =[]

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(pca_market_df)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
pca_elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow_pca = pd.DataFrame(pca_elbow_data)

df_elbow_pca.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
k_pca = df_elbow_pca.hvplot.line(
    x="k", 
    y="inertia", 
    title="PCA Elbow Curve", 
    xticks=k,
    label="PCA data"
)

k_pca
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** It appears 4 is the best value for 'k', as this is the point where the curve bends.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** This is the same best 'k' value found using original data.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialise the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the PCA data
model.fit(pca_market_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
market_pca_clusters = model.predict(pca_market_df)

# Print the resulting array of cluster values.
print(market_pca_clusters)
```

```python
# Create a copy of the DataFrame with the PCA data
pca_market_predictions = pca_market_df.copy()

# Add a new column to the DataFrame with the predicted clusters
pca_market_predictions["pca_market_clusters"] = market_pca_clusters

# Display sample data
pca_market_predictions.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Colour the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
predict_pca = pca_market_predictions.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="pca_market_clusters",
    hover_cols = ["coin_id"],
    label = "PCA data"
)

predict_pca
```
